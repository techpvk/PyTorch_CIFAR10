```python
import numpy as np
import yplot
from pathlib import Path
import yaml
import tqdm
import collections
from matplotlib import pyplot as plt
```

```python
logs_dir = Path('~/logs/selfens/concurrent_100epochs/configs').expanduser()
mlogs_dir = Path('~/logs/selfens/concurrent_250epochs/configs').expanduser()
```

```python
data = collections.defaultdict(dict)
mdata = collections.defaultdict(dict)
tdata = dict()
for config_dir in tqdm.tqdm(logs_dir.iterdir()):
    event_fn = next(config_dir.rglob('events.*'))
    hparams = yaml.load(next(config_dir.rglob('hparams.yaml')).open())
    events = yplot.read_tfevents(str(event_fn))
    try:
        acc = yplot.extract_scalar(events, 'accuracy/val')[-1]
        tacc = yplot.extract_scalar(events, 'teacher_accuracy/val')[-1]
        data[hparams['classifier']][hparams['num_students']] = acc
        tdata[hparams['classifier']] = tacc
    except:
        pass
for config_dir in tqdm.tqdm(mlogs_dir.iterdir()):
    event_fn = next(config_dir.rglob('events.*'))
    hparams = yaml.load(next(config_dir.rglob('hparams.yaml')).open())
    try:
        events = yplot.read_tfevents(str(event_fn))
        acc = yplot.extract_scalar(events, 'accuracy/val')[-1]
        mdata[hparams['classifier']][hparams['num_students']] = acc
    except:
        pass
```

```python
for arch in tdata:
    plt.figure()
    x, y = list(zip(*sorted(data[arch].items())))
    plt.plot(x, y, label='100 epochs')
    x, y = list(zip(*sorted(mdata[arch].items())))
    plt.plot(x, y, label='250 epochs')
    plt.plot(x, [tdata[arch]]*len(x), label='teacher')
    plt.title(arch)
    plt.legend()
    plt.xlabel('Number of Students')
    plt.ylabel('Validation Accuracy')
```

```python
import numpy as np
import yplot
from pathlib import Path
import yaml
import tqdm
import collections
from matplotlib import pyplot as plt
```

```python
logs_dir = Path('~/logs/selfens/async_250epochs_teachers/configs').expanduser()
logs_dir2 = Path('~/logs/selfens/async_250epochs_students/configs').expanduser()
```

```python
data = collections.defaultdict(dict)
mdata = collections.defaultdict(dict)
for config_dir in tqdm.tqdm(logs_dir.iterdir()):
    try:
        event_fn = next(config_dir.rglob('events.*'))
        hparams = yaml.load(next(config_dir.rglob('hparams.yaml')).open())
        events = yplot.read_tfevents(str(event_fn))
        acc = yplot.extract_scalar(events, 'accuracy/val')[-1]
        data[hparams['classifier']][hparams['num_eval_teachers']] = acc
    except:
        pass
for config_dir in tqdm.tqdm(logs_dir2.iterdir()):
    try:
        event_fn = next(config_dir.rglob('events.*'))
        hparams = yaml.load(next(config_dir.rglob('hparams.yaml')).open())
        events = yplot.read_tfevents(str(event_fn))
        acc = yplot.extract_scalar(events, 'accuracy/val')[-1]
        mdata[hparams['classifier']][hparams['num_eval_students']] = acc
    except:
        pass
```

```python
for arch in data:
    plt.figure()
    x, y = list(zip(*sorted(data[arch].items())))
    plt.plot(x, y, label='Teacher Ensemble')
    x, y = list(zip(*sorted(mdata[arch].items())))
    plt.plot(x, y, label='Student Ensemble')
    plt.title(arch)
    plt.legend()
    plt.xlabel('Number of Models')
    plt.ylabel('Validation Accuracy')
```
